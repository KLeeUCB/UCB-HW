```python
# matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

# File to Load (Remember to change these)
try:
    os.chdir(os.getcwd())
except:
    pass

city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_pd = pd.read_csv(city_data_to_load)
city_pd.head()

ride_pd = pd.read_csv(ride_data_to_load)
ride_pd.head()

# Combine the data into a single dataset
merge_data = pd.merge(city_pd, ride_pd, on="city", how="outer")

# Display the data table for preview
merge_data.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban = merge_data[merge_data['type'] == 'Urban']
suburban = merge_data[merge_data['type'] == 'Suburban']
rural = merge_data[merge_data['type'] == 'Rural']

urban_ride_count = urban.groupby(["city"]).count()["ride_id"]
urban_driver_count = urban.groupby(["city"]).mean()["driver_count"] 
urban_avg_fare = urban.groupby(["city"]).mean()["fare"]

suburban_ride_count = suburban.groupby(["city"]).count()["ride_id"]
suburban_driver_count = suburban.groupby(["city"]).mean()["driver_count"] 
suburban_avg_fare = suburban.groupby(["city"]).mean()["fare"]

rural_ride_count = rural.groupby(["city"]).count()["ride_id"]
rural_driver_count = rural.groupby(["city"]).mean()["driver_count"] 
rural_avg_fare = rural.groupby(["city"]).mean()["fare"]

# Build the scatter plots for each city types and incorporate a text label regarding circle size
plt.scatter(urban_ride_count, urban_avg_fare, s=10*urban_driver_count, c='coral', 
            edgecolor='black', linewidths=1, marker='o', label='Urban')

plt.scatter(suburban_ride_count, suburban_avg_fare, s=10*suburban_driver_count, c='skyblue', 
            edgecolor='black', linewidths=1, marker='o', label='Suburban')

plt.scatter(rural_ride_count, rural_avg_fare, s=10*rural_driver_count, c='gold', 
            edgecolor='black', linewidths=1, marker='o', label='Rural')

# Incorporate the other graph properties
x_limit = 40
x_axis = np.arange(0, x_limit, 1)
plt.ylim(18, 45)
plt.legend(loc="upper right")

# Create a legend
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")

# Save Figure
#plt.savefig("../Images/Pyber_Ride_Sharing_Data.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
type_counts = merge_data['type'].value_counts()
type_total = merge_data['type'].value_counts().sum()
type_total

urban_percent = urban / type_total
urban_percent

# # Build Pie Chart
plt.pie(urban_percent, autopct="%1.1f%%")
plt.axis()
plt.axis("equal")

# # Save Figure
# plt.savefig("../Images/total_fares.png")
```

## Total Rides by City Type

```python
# Calculate Ride Percents
urban_ride_count_total = urban.groupby(["city"]).count()["ride_id"].sum()

ride_percent = urban / urban_ride_count_total
ride_percent

# # Build Pie Chart
plt.pie(urban_percent, autopct="%1.1f%%")
plt.axis()
plt.axis("equal")

# # Save Figure
# plt.savefig("../Images/total_rides.png")
```

```python
# Show Figure
plt.show()
plt.axis("equal")
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
driver_total = merge_data['driver'].value_counts().sum()
driver_total

driver_percent = urban / driver_total
driver_percent

# Build Pie Charts
plt.pie(urban_percent, autopct="%1.1f%%")
plt.axis()
plt.axis("equal")

# Save Figure
```
